```python
import pandas as pd
import numpy as np

# 1. Load the data
data = pd.read_csv('FISHDTR.csv')
features = ['width', 'length']
output = 'weight'

# 2. Compute the correlation matrix
corr_matrix = data.corr().abs()

# 3. Extract the values for matrix C
C = np.zeros((3, 3))

for i in range(2):
    for j in range(2):
        C[i, j] = corr_matrix.iloc[i, j]

    C[i, 2] = corr_matrix.iloc[i, 2]
    C[2, i] = corr_matrix.iloc[2, i]

# Print matrix C
print(C)

# Determine which input feature is most correlated with the output
max_corr_index = np.argmax(C[:-1, -1])
strongest_feature = features[max_corr_index]
print(f"The input feature most strongly correlated with the output is: {strongest_feature}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Create a heatmap
labels = features + [output]
sns.heatmap(C, annot=True, cmap="YlGnBu", xticklabels=labels, yticklabels=labels)
plt.title("Correlation Matrix C")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('FISHDTR.csv')

# Fig. 5: Scatter-plot weight vs width
plt.figure(figsize=(8,6))
plt.scatter(data['width'], data['weight'], color='blue', alpha=0.5)
plt.title('Fig. 5: Weight vs Width')
plt.xlabel('Width')
plt.ylabel('Weight')
plt.grid(True)
plt.show()

```
